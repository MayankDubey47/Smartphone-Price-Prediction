# Linear Regression

```python
step1 = ColumnTransformer(transformers=[
    ('col_tnf',OneHotEncoder(sparse_output=False,drop='first'),[2,10,18]),
    ('col_tnf1', StandardScaler(),[4,5,6,8,9,11,12,13,16,17]),
    ('col_tnf2',OrdinalEncoder(categories=[['Single','Dual','Triple','Quad'],['0','Single','Dual'],['Blackberry','Android','Hongmeng','iOS'],['Quad Core','Hexa Core','Octa Core','Deca Core']]),['No_rear_camera','No_front_camera','OS','Num_Cores'])
],remainder='passthrough')

# step2 = 
step2 = LinearRegression()

pipe = Pipeline([
    ('step1',step1),
    
    ('step2',step2)
])

pipe.fit(X_train,y_train)

y_pred = pipe.predict(X_test)

print('R2 score',r2_score(y_test,y_pred))
print('MAE',mean_absolute_error(y_test,y_pred))
```

# Ridge

```python
step1 = ColumnTransformer(transformers=[
    ('col_tnf',OneHotEncoder(sparse_output=False,drop='first'),[2,10,18]),
    ('col_tnf1', StandardScaler(),[4,5,6,8,9,11,12,13,16,17]),
    ('col_tnf2',OrdinalEncoder(categories=[['Single','Dual','Triple','Quad'],['0','Single','Dual'],['Blackberry','Android','Hongmeng','iOS'],['Quad Core','Hexa Core','Octa Core','Deca Core']]),['No_rear_camera','No_front_camera','OS','Num_Cores'])
],remainder='passthrough')


step2 = Ridge(alpha=10)

pipe = Pipeline([
    ('step1',step1),
    
    ('step2',step2)
])

pipe.fit(X_train,y_train)

y_pred = pipe.predict(X_test)

print('R2 score',r2_score(y_test,y_pred))
print('MAE',mean_absolute_error(y_test,y_pred))
```

# Lasso Regression

```python
step1 = ColumnTransformer(transformers=[
    ('col_tnf',OneHotEncoder(sparse_output=False,drop='first'),[2,10,18]),
    ('col_tnf1', StandardScaler(),[4,5,6,8,9,11,12,13,16,17]),
    ('col_tnf2',OrdinalEncoder(categories=[['Single','Dual','Triple','Quad'],['0','Single','Dual'],['Blackberry','Android','Hongmeng','iOS'],['Quad Core','Hexa Core','Octa Core','Deca Core']]),['No_rear_camera','No_front_camera','OS','Num_Cores'])
],remainder='passthrough')

step2 = Lasso(alpha=0.001)

pipe = Pipeline([
    ('step1',step1),
    ('step2',step2)
])

pipe.fit(X_train,y_train)

y_pred = pipe.predict(X_test)

print('R2 score',r2_score(y_test,y_pred))
print('MAE',mean_absolute_error(y_test,y_pred))
```

# KNN

```python
step1 = ColumnTransformer(transformers=[
    ('col_tnf',OneHotEncoder(sparse_output=False,drop='first'),[2,10,18]),
    ('col_tnf1', StandardScaler(),[4,5,6,8,9,11,12,13,16,17]),
    ('col_tnf2',OrdinalEncoder(categories=[['Single','Dual','Triple','Quad'],['0','Single','Dual'],['Blackberry','Android','Hongmeng','iOS'],['Quad Core','Hexa Core','Octa Core','Deca Core']]),['No_rear_camera','No_front_camera','OS','Num_Cores'])
],remainder='passthrough')

step2 = KNeighborsRegressor(n_neighbors=3)

pipe = Pipeline([
    ('step1',step1),
    ('pca',PCA(n_components=0.95)),
    ('step2',step2)
])

pipe.fit(X_train,y_train)

y_pred = pipe.predict(X_test)

print('R2 score',r2_score(y_test,y_pred))
print('MAE',mean_absolute_error(y_test,y_pred))
```
